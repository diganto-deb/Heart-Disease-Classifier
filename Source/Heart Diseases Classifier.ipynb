```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('../Dataset/heart.csv')
```

```python

data.hist(figsize=(20,10))
plt.show()
```

```python
ss=StandardScaler()
```

```python
data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']]=ss.fit_transform(data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']])
```

```python
plt.figure(figsize=(10,5))
sb.countplot(data['target'])
plt.xlabel("Target Classes")
plt.ylabel("Count")
plt.title("Count of each Target Class")
plt.show()
```

```python
X=data.drop('target',axis=1)
y=data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Logistic Regression

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
l_pred=lr.predict(X_test)
```

```python
print(confusion_matrix(y_test,l_pred))
print('--'*40)
print(classification_report(y_test,l_pred))
```

## SVC

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
s_pred=svc.predict(X_test)
```

```python
print(confusion_matrix(y_test,s_pred))
print('--'*40)
print(classification_report(y_test,s_pred))
```

## KNN

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
fig=plt.figure(figsize=(20,10))
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(range(1,40),error_rate,color='blue', marker='o',markerfacecolor='red', linestyle='dashed', markersize=10)
axes.set_xlabel('K')
axes.set_ylabel('Error Rate')
axes.set_title('Error rate v/s K')
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=12)
```

```python
knn.fit(X_train,y_train)
k_pred=knn.predict(X_test)
print(confusion_matrix(y_test,k_pred))
print('\n')
print(classification_report(y_test,k_pred))
```

## Decision Tree 

```python
dtree=[]
for i in range(1,13):
    dtc=DecisionTreeClassifier(max_features=i)
    dtc.fit(X_train,y_train)
    pred_i=dtc.predict(X_test)
    dtree.append(np.mean(pred_i != y_test))
```

```python
fig=plt.figure(figsize=(20,10))
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(range(1,40),error_rate,color='green', marker='*',markerfacecolor='red', linestyle='dashed', markersize=10)
axes.set_xlabel('Max Features')
axes.set_ylabel('Scores')
axes.set_title('Scores v/s Max features')
plt.show()
```

```python
dtc=DecisionTreeClassifier(max_features=13)
```

```python
dtc.fit(X_train,y_train)
d_pred=dtc.predict(X_test)
```

```python
print(confusion_matrix(y_test,d_pred))
print('--'*40)
print(classification_report(y_test,d_pred))
```

## Random Forest

```python
rf_scores = []
estimators = [10, 100, 200, 500, 1000]
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i)
    rf_classifier.fit(X_train, y_train)
    rf_scores.append(rf_classifier.score(X_test, y_test))
    
```

```python
rf_scores=[round(i,2)for i in rf_scores]
```

```python
plt.figure(figsize=(10,5))
plt.bar([i for i in range(len(estimators))], rf_scores,  width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores[i], rf_scores[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators')
plt.show()
```

```python
rfc=RandomForestClassifier(n_estimators=500)
```

```python
rfc.fit(X_train,y_train)
r_pred=rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,r_pred))
print('--'*40)
print(classification_report(y_test,r_pred))
```
